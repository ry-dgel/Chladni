```python
%matplotlib inline
import matplotlib
matplotlib.use('Agg')
import numpy as np
from matplotlib import pyplot as plt
```

```python
radial_position_data_run1 = np.load('data/pos1.npy')
radial_position_data_run2 = np.load('data/pos2.npy')
angular_position_data_run1 = np.load('data/radpos1.npy')
angular_position_data_run2 = np.load('data/radpos2.npy')
```

First we do angular position. Let's plot these to get an idea:

```python
plt.plot(angular_position_data_run1)
plt.plot(angular_position_data_run2)
```

Let's fit the lowest point of the peak on each trial:

```python
np.argmin(angular_position_data_run1) - np.argmin(angular_position_data_run2)
```

So the points are about one angular position off. So let this be our error in angular position. Now, radial position:

```python
plt.plot(radial_position_data_run1)
plt.plot(radial_position_data_run2)
```

```python
np.argmin(radial_position_data_run1) - np.argmin(radial_position_data_run2)
```

So the points are about 107 points off. Let this be our error in radial positioning.
