```python
%matplotlib notebook
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
data_filename = '5_208_212_0.1.npy'
amplitude = np.load(data_filename)
summed_z = np.max(amplitude, 2)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
```

```python
grid_number = 5
dummy_axis = np.linspace(0, grid_number-1, grid_number)[:, None]
dummy_matrix = np.hstack((dummy_axis,)*grid_number).T
```

```python
ax.plot_surface(
    dummy_matrix, dummy_matrix.T, summed_z, cmap='pink_r', rstride=1, cstride=1)
```

```python
plt.show()
```
