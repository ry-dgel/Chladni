```python
%matplotlib notebook
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
data_filename = '17_1003_1009_0.2.npy'
amplitude = np.load(data_filename)
summed_z = np.sum(np.exp(amplitude), 2)
```

```python
plt.imshow(np.max(amplitude, 2))
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
```

```python
grid_number = 17
dummy_axis = np.linspace(0, grid_number-1, grid_number)[:, None]
dummy_matrix = np.hstack((dummy_axis,)*grid_number).T
```

```python
ax.plot_surface(
    dummy_matrix, dummy_matrix.T, np.max(amplitude, 2), cmap='pink_r', rstride=1, cstride=1)
```
